```python
import pandas as pd
import time
import bs4
import random
import os.path
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
custom = 'https://forms.gle/5wqwNd18WLnC16sNA'

chrome_options = Options()
# chrome_options.add_argument("--headless") # Ensure GUI is off
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument('disable-notifications')
chrome_options.add_argument('--disable-infobars')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument('window-size=1366x768')
# chrome_options.add_argument("start-maximized");
chrome_options.add_argument("--disable-extensions")
chrome_options.add_argument('--disable-blink-features=AutomationControlled')

# Set path to chromedriver as per your configuration
homedir = os.path.expanduser("~")
webdriver_service = Service(f"{homedir}/chromedriver/stable/chromedriver")

# Choose Chrome Browser
browser = webdriver.Chrome(service=webdriver_service, options=chrome_options)
browser.execute_script("document.body.style.zoom='75%'")


data = pd.read_excel('responden.xlsx', sheet_name='Email')
for idx in range(len(data)):
    emel = data.iloc[idx, 0]
    name = data.iloc[idx, 1]
    address = data.iloc[idx, 2]
    gender = data.iloc[idx, 3]
    print(name)
    # Get page
    browser = webdriver.Chrome(service=webdriver_service, options=chrome_options)
    browser.execute_script("document.body.style.zoom='75%'")
    browser.get(custom)

    try:
        email_p = WebDriverWait(browser, 10).until(EC.presence_of_element_located((By.XPATH, "/html/body/div[1]/div[2]/form/div[2]/div/div[2]/div/div/div/div[2]/div/div[1]/div/div[1]/input")))
    finally:
        time.sleep(3)
        email = browser.find_element(By.XPATH, "/html/body/div[1]/div[2]/form/div[2]/div/div[2]/div/div/div/div[2]/div/div[1]/div/div[1]/input")
        email.send_keys(emel) # looping sesuai data excel
        # description.submit()
        next1 = browser.find_element(By.XPATH, "/html/body/div[1]/div[2]/form/div[2]/div/div[3]/div[1]/div[1]/div/span")
        browser.execute_script("arguments[0].click();", next1)

    try:
        nama_p = WebDriverWait(browser, 10).until(EC.presence_of_element_located((By.XPATH, "/html/body/div/div[2]/form/div[2]/div/div[2]/div[2]/div/div/div[2]/div/div[1]/div/div[1]/input")))
        time.sleep(1)
    finally:
        nama = browser.find_element(By.XPATH, "/html/body/div/div[2]/form/div[2]/div/div[2]/div[2]/div/div/div[2]/div/div[1]/div/div[1]/input")
        nama.send_keys(name) # looping sesuai data excel
        male = "/html/body/div/div[2]/form/div[2]/div/div[2]/div[3]/div/div/div[2]/div/div/span/div/div[1]/label/div/div[1]/div/div[3]/div"
        female = "/html/body/div/div[2]/form/div[2]/div/div[2]/div[3]/div/div/div[2]/div/div/span/div/div[2]/label/div/div[1]/div/div[3]/div"
        if gender == 'L':
            path_gender = male
        elif gender == 'P':
            path_gender = female
        kelamin = browser.find_element(By.XPATH, path_gender)
        browser.execute_script("arguments[0].click();", kelamin)

        u24_25 = "/html/body/div/div[2]/form/div[2]/div/div[2]/div[4]/div/div/div[2]/div/div/span/div/div[3]/label/div/div[1]/div/div[3]/div"
        umur = browser.find_element(By.XPATH, u24_25)
        browser.execute_script("arguments[0].click();", umur)

        browser.execute_script("window.scrollTo(1, document.body.scrollHeight)")

        profes = ["Pegawai Negeri Sipil",
                  "Pegawai Swasta",
                  "Pelajar/Mahasiswa",
                  "Pengusaha/Wirausaha"]
        prof = random.choice(profes)
        idx_prof = profes.index(prof)+1
        work = browser.find_element(By.XPATH, f"/html/body/div/div[2]/form/div[2]/div/div[2]/div[5]/div/div/div[2]/div/div/span/div/div[{idx_prof}]/label/div/div[1]/div/div[3]/div")
        browser.execute_script("arguments[0].click();", work)

        sal = ["Rp 3.000.000 - Rp 5.000.000",
               "Rp 5.000.000 - Rp 10.000.000",
               "Rp 10.000.000 - Rp 15.000.000",
               "Rp 15.000.000 >"]
        sal_c = random.choice(sal)
        idx_sal = sal.index(sal_c)+1
        salary = browser.find_element(By.XPATH, f"/html/body/div/div[2]/form/div[2]/div/div[2]/div[6]/div/div/div[2]/div/div/span/div/div[{idx_sal}]/label/div/div[1]/div/div[3]/div")
        browser.execute_script("arguments[0].click();", salary)

        domisili = browser.find_element(By.XPATH, "/html/body/div/div[2]/form/div[2]/div/div[2]/div[7]/div/div/div[2]/div/div[1]/div/div[1]/input")
        domisili.send_keys(address) # looping sesuai excel

        next2 = browser.find_element(By.XPATH, "/html/body/div/div[2]/form/div[2]/div/div[3]/div[1]/div[1]/div[2]/span")
        browser.execute_script("arguments[0].click();", next2)
    try:
        edu_p = WebDriverWait(browser, 10).until(EC.presence_of_element_located((By.XPATH, "/html/body/div/div[2]/form/div[2]/div/div[2]/div[2]/div/div/div[2]/div/div/span/div/div[1]/label/div/div[1]/div/div[3]/div")))
        time.sleep(1)
    finally:
        edutech = browser.find_element(By.XPATH, "/html/body/div/div[2]/form/div[2]/div/div[2]/div[2]/div/div/div[2]/div/div/span/div/div[1]/label/div/div[1]/div/div[3]/div")
        browser.execute_script("arguments[0].click();", edutech)

        prod = ["Mini Class",
                "Webinar",
                "Bootcamp"]
        prod_l = prod.copy()
        # x = random.randint(1,3) # random jumlah pick
        x = 2
        print(f'jumlah pick : {x}')
        while x != 0:
            y = random.choice(prod_l) # memilih random produk
            idx_prod = prod.index(y)+1 # ambil index dr hasil pilihan lalu +1 agar cocok dengan xpath
            prod_l.remove(y) # hasil pilihan yang sudah diambil tidak boleh diambil lagi, maka dihapus
            x -= 1 # looping while agar sesuai dengan jumlah pilihan random yang diinginkan
            produk = browser.find_element(By.XPATH, f"/html/body/div/div[2]/form/div[2]/div/div[2]/div[3]/div/div/div[2]/div[1]/div[{idx_prod}]/label/div/div[1]/div[2]")
            browser.execute_script("arguments[0].click();", produk)

        br = ["Purwadhika",
              "Revou",
              "Hactive8",
              "Binar Academy",
              "Rakamin",
              "Myskill",
              "Dibimbing",
              "Digital Skola"]
        br_l = br.copy()
        # x = random.randint(1,3) # random jumlah pick
        x = 3
        print(f'jumlah pick : {x}')
        while x != 0:
            y = random.choice(br_l)
            idx_br = br.index(y)+1
            br_l.remove(y)
            x -= 1
            brand = browser.find_element(By.XPATH, f"/html/body/div/div[2]/form/div[2]/div/div[2]/div[4]/div/div/div[2]/div[1]/div[{idx_br}]/label/div/div[1]/div[2]")
            browser.execute_script("arguments[0].click();", brand)

        often_in = random.randint(1,5) # 5 karena cuma ada pilihan 1 2 3 4 >5
        often = browser.find_element(By.XPATH, f"/html/body/div/div[2]/form/div[2]/div/div[2]/div[5]/div/div/div[2]/div/div/span/div/div[{often_in}]/label/div/div[1]/div/div[3]/div")
        browser.execute_script("arguments[0].click();", often)

        genZ = browser.find_element(By.XPATH, f"/html/body/div/div[2]/form/div[2]/div/div[2]/div[6]/div/div/div[2]/div/div/span/div/div[1]/label/div/div[1]/div/div[3]/div")
        browser.execute_script("arguments[0].click();", genZ)

        next3 = browser.find_element(By.XPATH, "/html/body/div/div[2]/form/div[2]/div/div[3]/div[1]/div[1]/div[2]/span")
        browser.execute_script("arguments[0].click();", next3)

    try:
        nama_p = WebDriverWait(browser, 10).until(EC.presence_of_element_located((By.XPATH, "/html/body/div/div[2]/form/div[2]/div/div[3]/div[1]/div[1]/div[2]/span")))
        time.sleep(1)
    finally:
        next4 = browser.find_element(By.XPATH, "/html/body/div/div[2]/form/div[2]/div/div[3]/div[1]/div[1]/div[2]/span")
        browser.execute_script("arguments[0].click();", next4)


    def form(num_form: int,
             start_score: int,
             end_score: int,
             path_presence: str = '/html/body/div/div[2]/form/div[2]/div/div[2]/div[1]/div[2]',
             path_form: str = '/html/body/div/div[2]/form/div[2]/div/div[2]/div[{}]/div/div/div[2]/div/span/div/label[{}]/div[2]/div/div/div[3]/div',
             next_path: str = '/html/body/div/div[2]/form/div[2]/div/div[3]/div[1]/div[1]/div[2]/span'):
        '''
        path_presence : merupakan path pertama yang harus dilihat
        path_form : merupakan path dari form yang harus diisi
        num_form : jumlah kolom form sekaligus dengan keterangan
        next_path : path dari tombol next
        '''
        try:
            nama_p = WebDriverWait(browser, 10).until(EC.presence_of_element_located((By.XPATH, path_presence)))
            time.sleep(1)
        finally:
            for i in range(2, num_form+1):
                score_in = random.randint(start_score, end_score) # 5 karena cuma ada pilihan 3 4 5
                often = browser.find_element(By.XPATH, path_form.format(i, score_in))
                browser.execute_script("arguments[0].click();", often)

            next_p = browser.find_element(By.XPATH, next_path)
            browser.execute_script("arguments[0].click();", next_p)

    # Social Media Marketing
    socmed_marketing = form(6, 3, 5)

    # Sales Promotion
    ales_promotion = form(4, 4, 5)

    # Advertising
    advertising = form(5, 3, 5)

    # Perceived Value
    perceived_value = form(6, 3, 5)

    # Service Quality
    service_quality = form(6, 2, 4)

    # Electronic Word of Mouth
    word_mouth = form(6, 2, 4)

    # Purchase Decisions
    purchase_decision = form(6, 3, 5)

    send = browser.find_element(By.XPATH, "/html/body/div/div[2]/form/div[2]/div/div[3]/div[1]/div[1]/div[2]/span")
    browser.execute_script("arguments[0].click();", send)
    time.sleep(3)
    browser.quit()
    time.sleep(2)
    
```
